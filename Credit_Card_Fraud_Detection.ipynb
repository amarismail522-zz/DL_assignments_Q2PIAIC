
# Normalization of data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
data[['Time','Amount']]=sc.fit_transform(data[['Time','Amount']])
```

```python
data
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
x
```

```python
y
```

# Splitting dataset into Train and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
x.shape
```

```python
y.shape
```

```python
X_train,X_test,  Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state= 0 )
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
Y_train.shape
```

# Building a Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.regularizers import l2
```

```python
network=Sequential()
network.add(layers.Dense(10, activation='relu', kernel_regularizer=l2(0.001), input_shape=(X_train.shape[1],)))
network.add(layers.Dense(8, activation='relu', kernel_regularizer=l2(0.001)))
network.add(layers.Dense(1, activation='sigmoid'))
```

```python
network.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

```python
network_history = network.fit(X_train, Y_train, epochs=15, batch_size=350, validation_data=(X_val, Y_val))
```

# Visualizing Training and Validation Loss

```python
train_loss = network_history.history['loss']
```

```python
val_loss = network_history.history['val_loss']
```

```python
epochs = range(1,16)
```

```python
plt.plot(epochs, train_loss, 'r', label='Training Loss' )
plt.plot(epochs, val_loss, 'g', label='Validation Loss')
plt.title('Training and Validation loss Plot.')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
network.predict(X_test)
```

```python
network.evaluate(X_test, Y_test)
```
