```python
data.label.value_counts()
```

```python
#so here we will enconde the label because ML only know the Numerical value
replace_value={"label":{'g':0,'b':1}}
```

```python
data=data.replace(replace_value)
```

```python
#so we can see we encode the label 
data.head(15)
```

```python
data=data.sample(frac=1)
```

```python
data
```

```python
y=data.label
x=data.drop(columns="label")
```

```python
y.head()
```

```python
x
```

```python
#so data is ready now, we can split the data to training and testing
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size= 0.4 )
```

```python
X_train
```

```python
Y_train
```

```python
#Now we can build a model
from tensorflow.keras import Sequential, layers, regularizers
```

```python
network= Sequential()
network.add(layers.Dense(16, activation='relu', kernel_regularizer = regularizers.l2(0.001),  input_shape=(X_train.shape[1],)))
network.add(layers.Dense(8, activation='relu' , kernel_regularizer = regularizers.l2(0.001)))
network.add(layers.Dense(4, activation='relu' , kernel_regularizer = regularizers.l2(0.001)))
network.add(layers.Dense(1, activation='sigmoid'))
```

```python
network.compile(optimizer="adam",loss="binary_crossentropy", metrics="accuracy")
```

```python
network.fit(X_train, Y_train,epochs=70, batch_size=10)
```

```python
prediction = network.predict(X_test)
```

```python
network.evaluate(X_test, Y_test)
```
